As from the above the observations the data is highly unbalanced, and have very less proportion of fraudulent transanctions

Thus, I am deviding the datasets into two parts :

[0 -> Legit Transanctions] , [1 -> Fraud Transanctions]



```python
# Seperating the data from legit to fraud
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

**Under-Sampling **

Building the data set which would have similar distribution of fraudeulent transanctions and legit transanctions

```python
# No. of fraudeulent transanctions -> 492
# pinking around 492 random legit transanctions and putting them together in a data frame

legit_sample = legit.sample(n=492)
```

```python
# contantinating fraud and leget_sample
final_data = pd.concat([legit_sample, fraud], axis = 0)
final_data
```

```python
# Splitting the data into freatures and targets
X = final_data.drop(columns = 'Class', axis = 1)
Y = final_data['Class']
```

```python
# Splitting the data into training and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, stratify = Y, random_state = 101)
print(X.shape, X_train.shape, X_test.shape)
```

Training the model : Using Logistic regression

```python
model = LogisticRegression()
model
```

```python
model.fit(X_train, Y_train)
```

```python
# prompt: test it on the test data and print the acuracy score

y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)
```
